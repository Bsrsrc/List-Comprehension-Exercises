Using List Comprehension structure, convert the names of numeric variables in the car_crashes dataset to uppercase and prefix them with NUM_.

```python
df = sns.load_dataset("car_crashes")
df.head()
```

```python
df.info()
```

```python
["NUM_" + col.upper() if df[col].dtype != "O" else col.upper() for col in df.columns]
```

Using List Comprehension structure, append 'FLAG' to the names of variables in the car_crashes dataset that do not contain 'no' in their names.

```python
[col.upper() + "_FLAG" if "no" not in col else col.upper() for col in df.columns]
```

Using List Comprehension structure, select the names of variables that are DIFFERENT from the given variable names below and create a new dataframe.

```python
og_list = ["abbrev", "no_previous"]
new_cols = [col for col in df.columns if col not in og_list]
new_df = df[new_cols]
new_df.head()
```
